```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

G = nx.random_geometric_graph(200, .12)
```

```python
print G
```

```python
pos = nx.get_node_attributes(G, 'pos')
```

```python
dists = [(x - .5)**2 + (y - .5)**2 for x, y in list (pos.values())]
```

```python
ncenter = np.argmin(dists)
```

```python
p = nx.single_source_shortest_path_length(G, ncenter)
```

```python
plt.figure()
```

```python
nx.draw_networkx_edges(G, pos, alpha=.4)
nx.draw_networkx_nodes(G, pos, nodelist = list(p.keys()), node_size=120, alpha=.5, node_color=list(p.values()), cmap=plt.cm.jet_r)
```

```python
plt.show()
```
